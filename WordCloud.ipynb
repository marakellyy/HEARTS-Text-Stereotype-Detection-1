```python
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

df = pd.read_csv("immigrant_stereotype_dataset.csv")

# Clean text: remove URLs + lower-case
df["clean_text"] = df["text"].apply(lambda x: re.sub(r"http\S+", "", str(x)).lower())

# Update stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["https", "http", "co", "t", "rt"])

# Combine text
text = " ".join(df["clean_text"])

# Generate word cloud
wc = WordCloud(
    width=1200,
    height=600,
    background_color="white",
    stopwords=custom_stopwords,
    collocations=True
).generate(text)

plt.figure(figsize=(16, 8))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Adapted Immigrant Stereotype Dataset (Cleaned)", fontsize=18)
plt.show()

```
